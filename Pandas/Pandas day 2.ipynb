```python
data1.dropna(axis=1,thresh=3) #"cabin" column dont have atleast 3 values so it removed whole column
```

##### Q.Where PassengerId <3, no changes & PassengerId > 3, replace by logathmic

```python
import math

def cust1(x):
    if x<3:
        return x
    else:
        return math.log10(x)        
```

```python
data2['PassengerID_filter'] = data2['PassengerId'].apply(cust1)
data2
```
